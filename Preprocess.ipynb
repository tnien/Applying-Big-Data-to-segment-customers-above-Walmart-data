# Các bước tạo script

```python
from pyspark.sql.session import SparkSession

spark = SparkSession.builder.config('spark.driver.memory', '11g').getOrCreate()
```

```python
data_path = '/content/online_retail_II.csv'
from pyspark.sql.types import StructType, StructField, StringType, FloatType, DateType, IntegerType
from pyspark.sql.functions import to_date, regexp_replace, col
from pyspark.sql.functions import to_timestamp

# schema = StructType([
#     StructField("Invoice", StringType(), True),
#     StructField("StockCode", StringType(), True),
#     StructField("Description", StringType(), True),
#     StructField("Quantity", IntegerType(), True),
#     StructField("InvoiceDate", StringType(), True),
#     StructField("Price", StringType(), True),
#     StructField("CustomerID", IntegerType(), True),
#     StructField("Country", StringType(), True)
# ])
df = spark.read\
    .csv(data_path,
         header=True
        #  , schema=schema
         , inferSchema=True
         )
df = df.withColumn('InvoiceDate', to_timestamp('InvoiceDate', 'dd/MM/yyyy HH:mm'))

df.show(5)
```

```python
df.printSchema()
```

```python
df.count()
```

```python
df.describe().show()
```

```python
cancelledTransactions = df.filter(col("Invoice").rlike("C"))
```

## Biến đổi cột Price từ dấu , sang . để cast kiểu dữ liệu số

```python
df = df.withColumn("Price", regexp_replace(col("Price"), ",", ".").cast("float"))
```

```python
df.printSchema()
```

## Null values

```python
from pyspark.sql.functions import col
for col_name in df.columns:
  null_count = df.filter(col(col_name).isNull()).count()
  print(f"Null value count of column '{col_name}': {null_count}")
```

```python
df = df.na.drop(subset=['Invoice', 'Quantity', 'InvoiceDate', 'Price', 'Customer ID'])
df.count()
```

## Invalid value in **Price**

```python
df[df.Price <= 0].show(5)
```

### Đếm số hàng có giá trị âm hoặc bằng 0 ở cột Price

```python
df[~((df['Price'] > 0))].count()
```

### Loại bỏ những hàng có giá trị không phù hợp

```python
df = df[(df.Price >= 0)]
df.count()
```

## Duplicate value

```python
df.groupBy(df.columns).count().filter("count > 1").show()
```

```python
df = df.dropDuplicates()
```

```python
df.groupBy(df.columns).count().filter("count > 1").show()
```

## Drop Canceled Invoices

```python
df = df.filter(~col("Invoice").rlike("C"))
```
